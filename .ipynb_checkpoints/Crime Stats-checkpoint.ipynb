```python
import math
import warnings

from IPython.display import display
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from scipy.stats import mstats
from sklearn import linear_model
%matplotlib inline
pd.options.display.float_format = '{:.3f}'.format

# Suppress annoying harmless error.
warnings.filterwarnings(
    action="ignore",
    module="scipy",
    message="^internal gelsd"
)
crime = pd.read_csv(r'C:/Users/ezzan/Downloads/python_drills/police drill.csv')
```

```python
crime = crime[0:348]
crime = crime.fillna(value=0)
```

```python
crime
```

```python
crime1 = []
crime1 = crime[['Population']]
crime1['Murder'] = crime[['Murder and\nnonnegligent\nmanslaughter']]
crime1['Robbery'] = crime[['Robbery']]
crime1['Property_Crime'] = crime[['Property\ncrime']]
crime1
```

```python

crime1['Population'] = crime1.Population.str.replace(',','')
crime1['Property_Crime'] = crime1.Property_Crime.str.replace(',','')
crime1['Robbery'] = crime1.Robbery.str.replace(',','')
```

```python
crime1['Population'] = pd.to_numeric(crime1['Population'])
crime1['Robbery'] = pd.to_numeric(crime1['Robbery'])
crime['Murder'] = pd.to_numeric(crime1['Murder'])
crime['Property_Crime'] = pd.to_numeric(crime1['Property_Crime'])
```

```python
crime1['Population_Sqrd'] = np.square(crime1['Population'])
crime1['Murder'] = np.where(crime1['Murder'] >= 1, 1, 0)
crime1['Robbery'] = np.where(crime1['Robbery'] >= 1, 1, 0)
crime1
```

```python
crime2 = crime1.sort_values('Population',ascending= False)
crime2
```

```python
crime1 = crime1.drop(216)
crime1
```

```python
crime1["Property_Crime"] = mstats.winsorize(crime1["Property_Crime"], limits=0.05)
plt.figure(figsize=(10,10))
plt.scatter(crime1["Property_Crime"], crime1["Population"])
```

```python
crime1["Property_Crime"] = mstats.winsorize(crime1["Property_Crime"], limits=0.05)
plt.figure(figsize=(10,10))
plt.scatter(crime1["Property_Crime"], crime1["Population_Sqrd"])
```

```python
crime1.boxplot(column = 'Population')
```

```python
regr = linear_model.LinearRegression()
Y = crime1["Property_Crime"]
X = crime1[["Population", "Population_Sqrd", "Murder", "Robbery"]]
X = crime1[["Population", "Murder", "Robbery"]]
regr.fit(X, Y)

# Inspect the results.
print('\nCoefficients: \n', regr.coef_)
print('\nIntercept: \n', regr.intercept_)
print('\nR-squared:')
print(regr.score(X, Y))
```
